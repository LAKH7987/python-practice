```python
# if , else, elif
dist = int(input("enter the distance : "))
if dist > 10 :
  print("your charges bill is : 100/- ")
elif dist >= 7 and dist <= 10 :
  print("your charges bill is : 70/-")
elif dist >= 4 and dist < 7 :
    print("your charges bill is : 40/-")
else :
    print("your charges bill is : 30/-")
```
