```python
# 1. Take marks as input and print grade using if-elif ladder.

marks = int(input("enter your marks : "))
if marks > 90 :
    print("your grade is : A+")
elif marks <= 90 and marks >= 80 :
    print("your grade is : A")
elif marks < 80 and marks >= 65 :
    print("your grade is B+")
elif marks < 65 and marks >= 50 :
    print("your grade is : B")
elif marks < 50 and marks  >= 40 :
    print("your grade is : C")
elif marks < 40 and marks >=33 :
    print("your grade is D")
else :
    print("you have fail")
```

```python
# 7. Find common elements between two lists without using set intersection.

a = [1,2,3,7,7,9,0]
b = [5,8,0,4,7,6,9]
com = []
for i in a :
    if i in b and i not in com  :
        com.append(i)
print(com)
```

```python
# 13. Merge two dictionaries. If a key exists in both, sum their values.

d1 = {"a" : 1,"b" : 3,"c" : 5}
d2 = {"a" : 5,"b" : 9,"d" : 9}
c = {}
for i in d1 :
    c[i] = d1[i]
for i in d2 :
    if i in d1 :
        c[i] = c[i] + d2[i]
print(c)
```

```python
# 14. Create a function that returns factorial using loop.

def factorial(a) :
    fact = 1
    for i in range(1,a+1) :
        fact = fact * i 
    return fact
factorial(5)
```
