## Tuple


````python
# 3rd element extract
a = (10,20,30,40,50)
a[2]
````

````python
# last element with negatve indexing
a = (5,10,15,20)
a[-1]
     
````

````python
# extract all element using slicing
a = (100,200,300,4000,500)
a[:]
````

````python
# reverse tuple
a = (1,2,3,4,5)
a[::-1]
    
````

````python
# alternate element extract
a = (1,2,3,4,5,6)
a[::2]
    
````

````python
# extract first 3 elements
a = (11,22,33,44,55,66)
a[:3]
````

````python
# extract 2 to 4 elements from tuple
a = (1,2,3,4,5)
print(a[1:len(a)-1])
````

````python
# extract all elements from 3rd position
a = (5,10,15,20,25,30)
a[2:]
````

````python
# extract last 2nd element
a = (100,200,300,400)
a[-2]
````

````python
# extract first two and last two elements in tuple
a = (1,2,3,4,5,6)
b = []
b.extend(a[:2])
b.extend(a[-2:])
tuple(b)
````

````python
# sum of all elements of tuple
a = (1,2,3,4,5)
sum = 0
for i in a :
    sum += i
print("sum of all element is : ",sum)
````

````python
# count n=even number in tuple 
a = (1,2,3,4,5,6,6,78,8)
count = 0
for i in a :
    if i % 2 == 0 :
        count += 1
print(count)
````

````python
# multiply of all elements
a = (1,2,3,4,5)
multi = 1
for i in a :
    multi = multi*i
print("multiply of all number is : ",multi)
````

````python
# add element of two tuples
a = (1,2,3)
b =(4,5,6)
c = []
for i,j in zip(a,b) :
    sum = 0
    sum = (i+j)
    c.append(sum)
print(tuple(c))
    
````

````python
# extract all elements greater than 50
a = (20,55,60,30,75)
b = []
for i in a :
    if i > 50 :
        b.append(i)
print(tuple(b))
````

````python
# 15. Write a Python program to convert a string to a set of characters and remove all vowels.
# Input: `"datascience"`
# Output: Set of characters excluding vowels, e.g., `{'d', 't', 's', 'c', 'n'}`

str_ = "datascience"
b = set()
vowel = "AEIOUaeiou"
for i in str_ :
    if i not in vowel :
        b.update(i)
b
````
