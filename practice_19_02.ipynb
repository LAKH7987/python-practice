```python
# all prime number store in new list
a = [1,2,3,4,7,9,11,13,14,15]
new = []
for i in a :
    count = 0
    for j in range(1,i+1) :
        if i % j == 0 :
            count += 1
    if count == 2 :
        new.append(i)
    
print(new)
```

```python
# Find the second largest number in a list without using sort().

a = [-6,-5,-2,-2,-2,-3]
largest = a[0]
second = a[1]
if largest < second :
    largest,second = second,largest
for i in range(2,len(a)) :
    if a[i] != largest : 
        if a[i] > largest :
            second = largest
            largest = a[i]
        else : 
            if a[i] < largest and a[i] > second :
                second = a[i]
if largest == second :
    print("all numbers are equal")
else :
    print("largest : ",largest)
    print("second largest : ",second)
```

```python
# 10. Check if two strings are anagrams.

a = "earth"
b = "heart"
is_anagram = True

if len(a) == len(b) :
    for i in a :
        if i not in b :
            is_anagram = False 
            break
    if is_anagram :
        print("both strings are anagrams")
    else :
        print("strings are not anagrams")
else :
     print("strings are not anagrams")
```

```python
# 12. Write a program to find all pairs in a list whose sum equals a target number.

a = [1,2,3,4,6,5]
target = 9
l = len(a) 
for i in range(l) :
    for j in range(i+1,l) :
        if i != j :
            if a[i]+a[j] == target :
                print(a[i],a[j])
```

```python
# 15. Create a function that rotates a list by k positions.

a = [1,2,3,"no","hy",6,7,8,9]
k = 4
b = []
for i in range(k,len(a)) :
    b.append(a[i])
for i in range(k) :
    b.append(a[i])
print(b)
```
