# by using for and conditions

```python
# add '_' after each character
a = "python"
b = ""
for i in a :
    b += i
    b += "_"
print(b)
    
```

```python
# # check lower and uppercase
a = "i LOve PyThOn"
upper = 0
lower  = 0
for i in a :
    if ord(i) >= 65 and ord(i) <= 90 :
        upper += 1
    elif ord(i) >= 97 and ord(i) <= 122 :
          lower += 1
print("lowercase is : ",lower)
print("uppercase is : ",upper)
```

```python
# count character and digit
a = "abcdef123"
count = 0
ch = 0
for i in a :
    if ord(i) >= 48 and ord(i) <= 57 :
        count += 1
    elif (ord(i) >= 65 and ord(i) <= 90) or (ord(i) >= 97 and ord(i) <= 122) :
        ch += 1
print("character is : ",ch)
print("digit is : ",count)
```

```python
#10. Check if a given string is a palindrome (without slicing)

a = "madaam"
is_palindrome = True
i = 0
j = len(a)-1
while j > i :
    if a[i] != a[j] :
        is_palindrome = False
    i += 1
    j -= 1
if is_palindrome :
    print("string is palindrome ")
else :
    print("sting is non-palindrome")
    
```

```python
# Count the number of uppercase and lowercase letters in a string

a = "I Love PYthOn"
upper = 0
lower = 0
for ch in a :
    if ch.isupper() :
        upper += 1 
    if ch.islower() :
        lower += 1
print("uppercase is : ",upper)
print("lowercase is : ",lower)
```

```python
#3. Count the number of digits and alphabet characters in a string.

a = "python123j03"
digit = 0
alpha = 0
for i in a :
    if i.isdigit() :
        digit += 1
    if i.isalpha() :
        alpha += 1
print("digit : ",digit)
print("character is ",alpha)
```

```python
#4. Find the index of the first occurrence of a given character in a string.

a = "python"
b = a.index("t")
print(b)
```

```python
#5. Count how many times a given character appears in a string.

a = "i love python programming"
c = "o"
count = 0
for i in a :
    if c == i :
        count += 1
print(f"the occurence of {c} is : ",count)
```
