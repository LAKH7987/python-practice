```python
name1 = input("enter the first product name : ")
price1 = int(input("enter the price of product : "))
name2 = input("enter the second product name : ")
price2 = int(input("enter the price of product : "))
name3 = input("enter the third product name : ")
price3 = int(input("enter the price of product : "))
name4 = input("enter the fourth product name : ")
price4 = int(input("enter the price of product : "))
name5 = input("enter the fifth product name : ")
price5 = int(input("enter the price of product : "))

print()

print("          Billing System Of Mart ")
print("**...........................................**")
print()

print("Product Name        ---     Price of product")
print("**...........................................**")
print()
print(f"  {name1}                           {price1}")
print(f"  {name2}                           {price2}")
print(f"  {name3}                           {price3}")
print(f"  {name4}                           {price4}")
print(f"  {name5}                           {price5}")
print("**...........................................**")

add = price1+price2+price3+price4+price5

print()
print("total quantity :   5")
print(f"Bill of products : {add}")
print("**...........................................**")

gst = add*0.1
print("total gst =  ",gst)
print("Final bill after GST :")
print("**...........................................**")
add = gst + add
print()
print("total bill of Mart : ", add)


```
