```python
# Print a pyramid pattern:
#     *
#    ***
#   *****
#  *******

for i in range(1,6) :
    for j in range(5,i,-1) :
        print(" ",end="")
    for j in range(1,(2*i-1)+1) :
        print("*",end="")
    print()
```

```python
# Print alphabet pyramid:
#     1
 #   121
 #  12321
 # 1234321
 #123454321

for i in range(1,6) :
    for j in range(5,i,-1) :
        print(" ",end="")
    for k in range(1,i) :
        print(k,end="")
    for j in range(i,0,-1) :
        print(j,end="")
    print()
        
```

```python
# Write a program to print numbers from 1 to n, skipping numbers that are divisible by 3 or 5, but not both.

n = int(input("enter the number : "))

for i in range(1,n+1) :
    if i % 3 == 0 and i % 5 == 0 :
        print(i,end=" ")
        continue
    elif i % 3 == 0 :
        continue
    elif i % 5 == 0 :
        continue
    print(i,end=" ")

```
