```python
# Print a pyramid pattern:
#     *
#    ***
#   *****
#  *******

for i in range(1,6) :
    for j in range(5,i,-1) :
        print(" ",end="")
    for j in range(1,(2*i-1)+1) :
        print("*",end="")
    print()
```

```python
# Print alphabet pyramid:
#     1
 #   121
 #  12321
 # 1234321
 #123454321

for i in range(1,6) :
    for j in range(5,i,-1) :
        print(" ",end="")
    for k in range(1,i) :
        print(k,end="")
    for j in range(i,0,-1) :
        print(j,end="")
    print()
        
```

```python
# Write a program to print numbers from 1 to n but skip numbers whose digits sum is divisible by 4, and stop the loop when a palindrome number greater than 100 is encountered.

n = int(input("enter the number : "))

for i in range(1, n + 1):
    temp = i
    digit_sum = 0
    while temp > 0:
        r = temp % 10
        digit_sum += r
        temp //= 10

    divisible by 4
    if digit_sum % 4 == 0:
        continue

    # -------- palindrome check for i > 100 --------
    if i > 100:
        temp = i
        rev = 0
        while temp > 0:
            re =  temp % 10
            rev = rev * 10 + re
            temp //= 10

        if i == rev:
            break   # stop loop

    print(i, end=" ")
```
