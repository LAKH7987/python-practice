```python
# int -> float,string,complex,boolean

i = 10
print(i,type(i)) # int

f = float(i) # int -> float
print(f,type(f)) # float

s = str(i) # int -> string 
print(s,type(s))  # string

b = bool(i)  # int -> boolean
print(b,type(b)) # boolen

c = complex(i) # int -> complex
print(c,type(c)) # complex
```

```python
# float -> int,string,complex,boolean

f = 10.99
print(f,type(f)) # float

i = int(f) # float-> int
print(i,type(i)) # int

s = str(f) # float -> string 
print(s,type(s))  # string

b = bool(f)  # float-> boolean
print(b,type(b)) # boolen

c = complex(f) # float -> complex
print(c,type(c)) # complex
```

```python
# string -> int,float,complex,boolean

s = "123"
print(s,type(s)) # string

i = int(s) # string-> int
print(i,type(i)) # int

f = float(s) # string -> float 
print(f,type(f))  # float

b = bool(s)  # string -> boolean
print(b,type(b)) # boolen

c = complex(s) # string -> complex
print(c,type(c)) # complex
```

```python
# boolean -> int,float,complex,string

b = True 
print(b,type(b)) # string

i = int(b) # boolean-> int
print(i,type(i)) # int

f = float(b) # boolean -> float 
print(f,type(f))  # float

s = str(b)  # boolean-> string
print(s,type(s)) # string

c = complex(b) # boolean -> complex
print(c,type(c)) # complex
```

```python
# complex -> int,float,boolean,string

c = 2+3j 
print(c,type(c)) # complex

i = int(c.real) # complex-> int
print(i,type(i)) # int real number

i = int(c.imag) # complex-> int
print(i,type(i)) # int imaginary number  

f = float(c.real) # complex -> float 
print(f,type(f))  # float real number

f = float(c.imag) # complex -> float 
print(f,type(f))  # float imaginary number

s = str(b)  # complex-> string
print(s,type(s)) # string

b = complex(c) # complex -> boolean
print(b,type(b)) # boolean
```
